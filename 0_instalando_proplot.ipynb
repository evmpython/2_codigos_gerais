```python
# cria a moldura da figura
fig, ax = pplt.subplots(figsize=(5,3),
                        tight=True,
                        proj='pcarree')

# formatação dos eixos
ax.format(coast=True, borders=True, innerborders=False,
          labels=True, latlines=10, lonlines=10,
          title='Mapa Mundi')

# exibe a figura
pplt.show()
```
