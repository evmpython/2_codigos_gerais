```python
# Use an on-the-fly projection
import ultraplot as uplt
import warnings
warnings.filterwarnings("ignore")

fig = uplt.figure(refwidth=3)
axs = fig.subplots(nrows=2, proj="robin", proj_kw={"lon0": 150})

# proj = uplt.Proj('robin', lon0=180)
# axs = uplt.subplots(nrows=2, proj=proj)  # equivalent to above

axs.format(suptitle="Figure with single projection",
           land=True,
           latlines=30,
           lonlines=60)
```

```python
# Simple subplot grid
import numpy as np
import ultraplot as uplt

state = np.random.RandomState(51423)
data = 2 * (state.rand(100, 5) - 0.5).cumsum(axis=0)

fig = uplt.figure()
ax = fig.subplot(121)
ax.plot(data, lw=2)

ax = fig.subplot(122)

fig.format(suptitle="Simple subplot grid", title="Title", xlabel="x axis", ylabel="y axis")
```
