Lista os arquivos

```python
files = sorted(glob.glob(f'{dir}input/Historico*csv*'))
files
```

Concatena dataframes

```python
# cria uma tabela (conhecida também com o nome em inglês de "dataframe") vazia
df = pd.DataFrame()

# loop de cada arquivo da lista files
for file in files:

    # leitura da tabela
    df0 = pd.read_csv(file, delimiter=';')

    # junta a tabela que foi lida com a anterior
    df = pd.concat([df, df0], ignore_index=True)

    # imprime o nome do arquivo lido na tela
    print(file, df0.shape[0])

# mostra o dataframe
df
```

Transdorma a coluna "Data" para o índice do dataframe

```python
# transformando a coluna "Data" para datetime
df['Data'] = pd.to_datetime(df['Data']) # 18/02/2023 17:31:47

# coluna data como índice da tabela
df.set_index('Data', inplace=True)

# mostra o dataframe
df
```

Seleciona os dados que estão num intervalo de data

```python
# data central
ano, mes, dia, hor, min = '2023', '02', '18', '17', '40'

# Data inicial e final em formato de "CARACTER"
data_str_ini = str(datetime(int(ano), int(mes), int(dia), int(hor), int(min), 0, 1))                                                # tempo inicial
data_str_end = str(datetime.strptime(data_str_ini, '%Y-%m-%d %H:%M:%S.%f') + timedelta(minutes=4, seconds=59, microseconds=999998)) # tempo final (+5min)

print(data_str_ini)
print(data_str_end)

# procura na tabela da Brasildat, relâmpagos dentro do tempo do relâmpago da SPLMA
df_recorte = df.loc[data_str_ini : data_str_end]
```
