# Usando `111.195`

```python
# distância máxima entre a estação e o foco de calor
dismax = 20.

# importa bibliotecas
import numpy as np

# cria uma cópia explícita para evitar o warning
df_trabalho = df_2024.copy()

for indice, linha in df_estacoes_selecionadas.iterrows():

    # lat, lon e nome da cidade
    late, lone, munie = linha['Lat'], linha['Lon'], linha['Estação']

    # calcula distância
    dis = (np.sqrt((df_trabalho['lat'].values - late)**2 + (df_trabalho['lon'].values - lone)**2)) * 111.195

    # usando .loc para atribuir valores
    df_trabalho.loc[:, 'distancia_km'] = dis

    # insere a distância no dataframe
    df_focos_estacao = df_trabalho[df_trabalho['distancia_km'] <= dismax]

    # mostra na tela a qte de focos por cidade
    print(munie, df_focos_estacao.shape[0])

    # salva em CSV
    if df_focos_estacao.shape[0] > 0:
        df_focos_estacao.to_csv(f"/content/08_focoscalor_por_estacao/{munie}_{str(int(dismax))}km.csv")
```
