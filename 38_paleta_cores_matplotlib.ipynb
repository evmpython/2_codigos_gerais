```python
from matplotlib.colors import LinearSegmentedColormap
cmap = LinearSegmentedColormap.from_list('azul_vermelho', ['blue', 'cyan', 'green', 'yellow', 'orange', 'red', 'gray'], N=18)
cores_18 = [cmap(i) for i in range(18)]  # Cores no formato RGBA
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 2))

for i, cor in enumerate(cores_18):
    plt.bar(i, 1, color=cor, edgecolor='black')

plt.xticks([])
plt.yticks([])

plt.title('18 Cores Interpoladas (Azul â†’ Vermelho)')

plt.show()
```
