```python
import matplotlib.pyplot as plt

# Plot all Brazilian states
fig, ax = plt.subplots(figsize=(4,4), dpi=300)

bacia_rio_paraíba_do_sul.plot(facecolor="#2D3E50", edgecolor="#FEBF57", ax=ax)

ax.set_title('Bacia do Rio Paraíba do Sul', fontsize=7)
ax.axis("off")
```
